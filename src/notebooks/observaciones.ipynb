## Resumen del Data Cleaning

### Dataset final
- **Registros**: 498
- **Columnas**: 14
- **Período**: Octubre — Diciembre 2025
- **Organismos**: AGENCIA TRIBUTARIA y RECAUDACIÓN TRIBUTARIA

### Columnas finales

| # | Columna | Tipo | Descripción |
|---|---------|------|-------------|
| 0 | referencia | str | ID único de la subasta |
| 1 | organismo | str | Entidad convocante |
| 2 | estado | str | Estado de la subasta |
| 3 | tipo_subasta | str | AEAT o Recaudación Tributaria |
| 4 | fecha_inicio | datetime | Fecha de inicio de la subasta |
| 5 | fecha_conclusion | datetime | Fecha de conclusión |
| 6 | lotes | int | Número de lotes |
| 7 | valor_subasta_eur | float | Precio de salida (€) |
| 8 | tasacion_eur | float | Valor de mercado (€) |
| 9 | puja_minima_eur | float | Oferta mínima aceptada (€) |
| 10 | tramo_pujas_eur | float | Incremento mínimo entre pujas (€) |
| 11 | deposito_eur | float | Fianza previa 5% (€) |
| 12 | descuento_pct | float | Descuento sobre tasación (%) |
| 13 | duracion_dias | float | Duración de la subasta en días |

### Decisiones tomadas

**Columnas eliminadas:**
- `id_sub`, `referencia_det`, `num_lotes_listado` → duplicadas
- `url_detalle`, `anuncio_boe` → no aportan análisis
- `fecha_conclusion_listado` → formato sucio, sustituida por `fecha_conclusion`
- `forma_adjudicacion`, `expediente`, `cuenta_expediente` → solo judicial
- `descripcion` → texto libre, no aporta al análisis numérico
- `cantidad_reclamada_eur` → solo judicial, no aplica a AEAT

**Nulos justificados:**
- `fecha_inicio`, `fecha_conclusion`, `duracion_dias` → 28 subastas canceladas
- `puja_minima_eur` → "Sin puja mínima" es información válida

**Columnas calculadas:**
- `descuento_pct` = (tasacion - valor_subasta) / tasacion × 100
- `duracion_dias` = fecha_conclusion - fecha_inicio
